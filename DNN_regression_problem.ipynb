```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_percentage_error,mean_squared_error,r2_score

import warnings
warnings.filterwarnings('ignore')
```

```python
# basic package of DNN
import tensorflow as tf
from tensorflow import keras
```

```python
from sklearn.datasets import load_boston
x,y= load_boston(return_X_y=True)
x= pd.DataFrame(x)
y= pd.DataFrame(y)
x.isna().sum()
```

```python
sc_x = StandardScaler().fit_transform(x)
x= pd.DataFrame(sc_x)
x.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=101)
xtrain.shape
```

## Deep Learning- Neural Network -- Multi Layer Perceptron

API

Sequential api

functional api

### Sequential API tp build DNN

```python
DNN = tf.keras.models.Sequential()

# Add first layer with input
DNN.add(tf.keras.layers.Dense(units=6,activation='relu'))

# Add first layer to second layer
DNN.add(tf.keras.layers.Dense(units=9,activation='relu'))

# Add second layer with third layer
DNN.add(tf.keras.layers.Dense(units=7,activation='relu'))

# Add third layer with output layer
DNN.add(tf.keras.layers.Dense(units=1))
```

```python
# Apply Optimization Techniques
DNN.compile(optimizer='adam',loss='mean_squared_error')
```

```python
# apply input dataset along with backpropogation
DNN.fit(xtrain,ytrain,epochs=299,batch_size=30)
```

```python
y_pred = DNN.predict(xtest,batch_size=29)
```

```python
r2_score(ytest,y_pred)
```
